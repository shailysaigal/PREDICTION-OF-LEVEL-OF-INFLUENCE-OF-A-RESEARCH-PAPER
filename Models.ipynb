```python
import numpy as np
import pandas as pd
import matplotlib
```

```python
data = pd.read_excel('final2.xlsx',thousands = ',')
```

```python
data= pd.DataFrame(data)
#from sklearn.utils import shuffle
#data = data.sample(frac=1).reset_index(drop=True)
data.head(10)
```

```python
#X = data.drop(['Rank','Papers','Specificity','Orientation','Under_Implication_Contributuion_Section','Cited_in_different_section','A11','A16','A17','A21'],axis=1)
X = data.drop(['Rank','Papers','Year','Under_the_Discussion_section','Part_of_a_cluster_ of_citation',
               'Pubjished_in_the_same_journal', 'Total_Number_of_Authors',
               'A5','A16','A17','A20','Total_years'],axis=1)
```

```python
import statsmodels.api as sm
X = sm.add_constant(X)
Y = data['Rank'] 
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
X_train = X_train.select_dtypes(include=[np.number])
X_test = X_test.select_dtypes(include=[np.number])

num_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy="median")),
        ('std_scaler', StandardScaler()),
    ])

X_train_Scalar = num_pipeline.fit_transform(X_train)
print(X_train_Scalar)

X_test_Scalar = num_pipeline.fit_transform(X_test)
print(X_test_Scalar)

print(X_train_Scalar.shape,X_test_Scalar.shape)
print(Y_train.shape,Y_test.shape)

X_train=np.concatenate([X_train_Scalar],axis=1)
X_train=pd.DataFrame(X_train)

X_test=np.concatenate([X_test_Scalar],axis=1)
X_test=pd.DataFrame(X_test)
print(type(X_train),type(X_test))
```

```python
np.isnan(X_train).any(), np.isnan(Y_train).any(), np.isnan(X_test).any()
X_train[np.isnan(X_train)] = np.median(X_train[~np.isnan(X_train)])
Y_train[np.isnan(Y_train)] = np.median(Y_train[~np.isnan(Y_train)])
Y_test[np.isnan(Y_test)] = np.median(Y_test[~np.isnan(Y_test)])
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(solver="liblinear", random_state=42)
# same for solver='newton-cg'

log_predict=log_reg.fit(X_train, Y_train).predict(X_test)
#print(log_predict)

from sklearn.metrics import accuracy_score
print("Accuracy: ",accuracy_score(Y_test, log_predict))
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_test, log_predict)
from sklearn.metrics import classification_report
print(classification_report(Y_test, log_predict))
from sklearn.metrics import mean_squared_error
rmse = mean_squared_error(Y_test, log_predict)
print("RMSE:",rmse)
from sklearn.metrics import confusion_matrix
confusion = confusion_matrix(Y_test, log_predict)
print("Confusion matrix:\n",confusion)
```

```python
from sklearn.svm import SVC
svr_rbf = SVC(kernel='rbf', C=4,gamma='scale',random_state=42)
y_rbf = svr_rbf.fit(X_train, Y_train).predict(X_test)

print("Accuracy: ",accuracy_score(Y_test, y_rbf))
confusion_matrix(Y_test, y_rbf)
print(classification_report(Y_test, y_rbf))
rmse = mean_squared_error(Y_test, y_rbf)
print("RMSE:",rmse)
confusion = confusion_matrix(Y_test, y_rbf)
print("Confusion matrix:\n",confusion)
```

```python
from sklearn.svm import SVC
svr = SVC(kernel='linear', C=0.5,gamma='scale',random_state=42)
y_lin = svr.fit(X_train, Y_train).predict(X_test)

print("Accuracy: ",accuracy_score(Y_test, y_lin))
confusion_matrix(Y_test, y_lin)
print(classification_report(Y_test, y_lin))
rmse = mean_squared_error(Y_test, y_lin)
print("RMSE:",rmse)
confusion = confusion_matrix(Y_test, y_lin)
print("Confusion matrix:\n",confusion)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_clf = DecisionTreeClassifier(max_depth=5, random_state=42,min_samples_leaf=3)
dec_tree=tree_clf.fit(X_train, Y_train).predict(X_test)

print("Accuracy: ",accuracy_score(Y_test, dec_tree))
confusion_matrix(Y_test,dec_tree)
print(classification_report(Y_test, dec_tree))
rmse = mean_squared_error(Y_test, dec_tree)
print("RMSE:",rmse)
confusion = confusion_matrix(Y_test, dec_tree)
print("Confusion matrix:\n",confusion)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier

log_clf = LogisticRegression(solver="liblinear", random_state=42)
rnd_clf = RandomForestClassifier(n_estimators=6, random_state=42)
svm_clf = SVC(gamma="auto", random_state=42,probability=True)

voting_clf = VotingClassifier(
    estimators=[ ('lr', log_clf),('rf', rnd_clf), ('svc', svm_clf)],
    voting='hard')

voting_clf.fit(X_train, Y_train)
for clf in (log_clf, svm_clf,rnd_clf, voting_clf):
    clf.fit(X_train, Y_train)
    y_pred = clf.predict(X_test)
    print(clf.__class__.__name__, accuracy_score(Y_test, y_pred))
    confusion_matrix(Y_test,y_pred)
    print(classification_report(Y_test, y_pred))
    rmse = mean_squared_error(Y_test, y_pred)
    print("RMSE:",rmse)
    confusion = confusion_matrix(Y_test, y_pred)
    print("Confusion matrix:\n",confusion)
    print("\n")
   
```

```python
from sklearn.ensemble import BaggingClassifier

bag_clf = BaggingClassifier(
    DecisionTreeClassifier(splitter="random", max_leaf_nodes=16,random_state=42), n_estimators=500,
    max_samples=1.0, random_state=42)

bag_clf.fit(X_train, Y_train)
y_pred_bag = bag_clf.predict(X_test)
print("Accuracy: ",accuracy_score(Y_test, y_pred_bag))
confusion_matrix(Y_test,y_pred_bag)
print(classification_report(Y_test, y_pred_bag))
rmse = mean_squared_error(Y_test, y_pred_bag)
print("RMSE:",rmse)
confusion = confusion_matrix(Y_test, y_pred_bag)
print("Confusion matrix:\n",confusion)
```
