```python
#X = data.drop(['Rank'], axis=1)
X = data.drop(['Rank','Papers','Year'], axis=1)
from sklearn import preprocessing
Y = data['Rank']
print(X.columns)
X.head()
```

```python
np.random.seed(42)
#from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
#stratify=X['Total_Number_of_Authors'] gives error
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
X_train = X_train.select_dtypes(include=[np.number])
X_test = X_test.select_dtypes(include=[np.number])

num_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy="median")),
        ('std_scaler', StandardScaler()),
    ])

X_train_Scalar = num_pipeline.fit_transform(X_train)
print(X_train_Scalar)

X_test_Scalar = num_pipeline.fit_transform(X_test)
print(X_test_Scalar)

print(X_train_Scalar.shape,X_test_Scalar.shape)
print(Y_train.shape,Y_test.shape)
```

```python
X_train=np.concatenate([X_train_Scalar],axis=1)
X_train=pd.DataFrame(X_train)

X_test=np.concatenate([X_test_Scalar],axis=1)
X_test=pd.DataFrame(X_test)
print(type(X_train),type(X_test))
```

```python
np.isnan(X_train).any(), np.isnan(Y_train).any(), np.isnan(X_test).any()
X_train[np.isnan(X_train)] = np.median(X_train[~np.isnan(X_train)])
Y_train[np.isnan(Y_train)] = np.median(Y_train[~np.isnan(Y_train)])
Y_test[np.isnan(Y_test)] = np.median(Y_test[~np.isnan(Y_test)])
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
#Y = Y_train
#Y = pd.DataFrame(Y)
Y = preprocessing.scale(Y_train)
Y = pd.DataFrame(Y)

#X_scaled = preprocessing.scale(X_train)
X=X_train

import statsmodels.api as sm
#X = pd.DataFrame(X_scaled, columns = X_train.columns)
#X=X_train
X_train = sm.add_constant(X_train)
lasso= Lasso()
parameters = {'alpha': [1e-15,1e-10,1e-8,1e-4,1e-3,1e-2,1,5,10,20,30,40,100]}
lasso_regGrid = GridSearchCV(lasso,parameters)
lasso_regGrid.fit(X, Y)
print(lasso_regGrid.best_params_)

lasso_reg = Lasso(alpha=0.01)
lasso_reg.fit(X, Y)
coef=lasso_reg.coef_
print(coef)
```
